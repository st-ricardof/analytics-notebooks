```python
import requests ##biblioteca que faz requisições

parametros = {
    'ano' : 2021,
    'codcur': ''
}

resultado = requests.get(url = 'https://dados.fflch.usp.br/api/defesas', params = parametros)
dados = resultado.json() ##indicando de que tipo é o dado

print(dados)
```

```python
##transformar os dados em um formato tabular e para isso usamos o pandas
import pandas as pd

tabela = pd.DataFrame(dados)

tabela.head() ##exibe o começo da tabela
```

```python
import re 
cursos = tabela['nomcur'].value_counts()
defesasPorCurso = {}
for i,v in cursos.items():
  curso = re.sub('\s\([^()]*\)', "", i )
  curso = 'História' if 'História' in curso else curso 
  curso = 'Letras' if any(substr in curso for substr in ['Tradução', 'Árabes', 'Língua', 'Lingüística', 'Literatura', 'Letras']) else curso#It will return True if any of the substrings in substring_list is contained in string.
  curso = 'Ciência Social' if any(substr in curso for substr in ['Política', 'Sociologia']) else curso
  try:
    defesasPorCurso[curso] += v
  except KeyError:
    defesasPorCurso[curso] = v
print(cursos)
defesasPorCurso
```
