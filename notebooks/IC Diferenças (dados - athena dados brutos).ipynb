```python
!pip install xlrd
import pandas as pd
import requests
```

```python
com = requests.get("https://dados.fflch.usp.br/api/pesquisa/iniciacao_cientifica?departamento=FLF&curso=&bolsa=true&export=false&tipo=anual&ano=2021&ano_ini=&ano_fim=")
sem = requests.get('https://dados.fflch.usp.br/api/pesquisa/iniciacao_cientifica?departamento=FLF&curso=&bolsa=false&export=false&tipo=anual&ano=2021&ano_ini=&ano_fim=')
```

```python
bolsa = pd.DataFrame(com.json())
sem_bolsa = pd.DataFrame(sem.json())
```

```python
df1 = pd.concat([bolsa, sem_bolsa], ignore_index=True)
```

```python
path = '/home/ricardo/Downloads/Projetos_IC_2021 - Modelo Atena.xls'
```

```python
df2 = pd.read_excel(path)
df2 = df2.loc[df2['DEPARTAMENTO'] == 'Filosofia']
```

```python
merged = pd.merge(df1, df2, how='outer', left_on=['codproj', 'ano_proj'], right_on=['COD_PROJETO', 'ANO_PROJETO'])
```

```python
column_names = ['ano_proj','ANO_PROJETO', 'codproj', 'COD_PROJETO', 'nome_discente', 'ALUNO', 'genero_discente',
       'raca_cor_discente', 'nome_supervisor', 'ORIENTADOR', 'genero_supervisor',
       'titulo_pesquisa',  'TITULO', 'data_ini', 'DATA_INICIO_PROJETO',  'data_fim','DATA_FIM_PROJETO',  'bolsa', 'status_projeto', 'SITUAÇÃO_PROJETO',
       'SITUAÇÃO_BOLSA_PIBI', 'OUTRO FOMENTO',
       'dtainibol', 'dtafimbol', 'sigla_departamento', 'nome_departamento','cod_curso', 'nome_curso', 'DEPARTAMENTO', 'id']

merged = merged.reindex(columns=column_names)
```

```python
df = merged.drop(['genero_discente','raca_cor_discente','genero_supervisor','sigla_departamento', 'nome_departamento','cod_curso', 'nome_curso', 'DEPARTAMENTO', 'id'], axis=1)
```
