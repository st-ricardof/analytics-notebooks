```python
# !pip install xlrd #read xls
# !pip install openpyxl #read xlsx
import pandas as pd
import requests
import time
```

```python
portal_path = "http://127.0.0.1:8001/api/pesquisa/pos_doutorandos"
cpq_path = '/home/ricardo/Downloads/Projetos_PD_TipoApoio 2017-2022 - completa.xlsx - Projetos_PD_TipoApoio.xls.csv'
```

```python
listadpt = ['FLA', 'FLP', 'FLF', 'FLH', 'FLC', 'FLM', 'FLO', 'FLL', 'FSL', 'FLT', 'FLG']

listadptext = ['Antropologia', 'Ciência Política', 'Filosofia', 'História', 'Letras Clássicas e Vernáculas', 
                'Letras Modernas', 'Letras Orientais', 'Lingüística', 'Sociologia', 
                'Teoria Literária e Literatura Comparada', 'Geografia']
```

```python
dfDiff = pd.DataFrame(columns={'codproj', 'Código', 'ano_proj', 'Ano ', 'nome_discente', 
                'Pósdoutorando', 'status_projeto', 'Situação', 'titulo_pesquisa', 
                'titulo_pesquisa', 'nome_departamento', 'Departamento '}) 

for dpt, dptext in zip(listadpt, listadptext):
    for ano in range(2017, 2022):
        parametros = {
        'ano': ano,
        'departamento': dpt,
        'departamento_ext': dptext,
        'export': 'false',
        'tipo': 'anual'
        }

        time.sleep(2)
        com = requests.get(portal_path+"?bolsa=true", params= parametros)
        time.sleep(1)
        sem = requests.get(portal_path+"?bolsa=false", params= parametros)

        try:    
            bolsa = pd.DataFrame(com.json())
            sem_bolsa = pd.DataFrame(sem.json())
        except:
            print(parametros)
            print(com.status_code) 
            print('\n')
            continue

        portal = pd.concat([bolsa, sem_bolsa], ignore_index=True)

    #        COLUNAS PORTAL
    #        Index(['id', 'codproj', 'nome_discente', 'nome_supervisor', 'titulo_pesquisa',
    #        'data_ini', 'data_fim', 'bolsa', 'apoio_financeiro', 'agencia_fomento',
    #        'dtainibol', 'dtafimbol', 'obs', 'sigla_departamento',
    #        'nome_departamento', 'cod_curso', 'nome_curso'],
    #        dtype='object')

        #cpq = pd.read_excel(cpq_path, sheet_name='PosDoc')
        cpq = pd.read_csv(cpq_path)#exporeti para .csv, pois estava dando um warning em relação à extensão do microsft excel

        cpq = cpq.loc[(cpq['Departamento Projeto'] == parametros['departamento_ext']) & (cpq['Ano Projeto'] == str(parametros['ano'])) ]
        cpq = cpq.reset_index(drop=True)

        portal = portal[['codproj', 'ano_proj', 'nome_discente', 'titulo_pesquisa', 'status_projeto', 'nome_departamento']]
        cpq =  cpq[['Código Projeto', 'Ano Projeto', 'Nome Pós-Doutorando', 'Titulo Projeto', 'Situação Atual Projeto', 'Departamento Projeto']]

#         COLUNAS CPQ
#         Index(['Ano Projeto', 'Código Projeto', 'Situação Atual Projeto',
#        'Número USP Pós-Doutorando', 'Nome Pós-Doutorando',
#        'Email Pós-Doutorando', 'Número USP Proponente', 'Nome Proponente',
#        'Email Proponente', 'Departamento Projeto', 'Titulo Projeto',
#        'Último Apoio Projeto',
#        'Última Agência Fomento/Instituição Afastamento', 'Data Início Projeto',
#        'Data Fim Projeto'],
#       dtype='object')
        

        cpq = cpq.astype({'Código Projeto' : str, 'Ano Projeto' : str})
        portal = portal.astype({'codproj' : str, 'ano_proj' : str})

        df = pd.merge(
        portal, 
        cpq, 
        how='outer', 
        left_on=['codproj', 'ano_proj'], 
        right_on=['Código Projeto', 'Ano Projeto']
                        )

        column_names = ['codproj','Código Projeto', 'ano_proj', 'Ano Projeto', 'nome_discente', 
            'Nome Pós-Doutorando', 'status_projeto', 'Situação Atual Projeto', 'titulo_pesquisa', 
            'Titulo Projeto', 'nome_departamento', 'Departamento Projeto']

        df = df.reindex(columns=column_names)
        df = df.sort_values(by=['Nome Pós-Doutorando'], ascending=True)
        df = df.reset_index(drop=True)

        df = df.loc[df['nome_discente'].isnull() | df['Nome Pós-Doutorando'].isnull()]

        #dfDiff=dfDiff.append( [df] ) # deprecated
        dfDiff = pd.concat([dfDiff, df])
```
