```python
import requests 
import json

with open('/content/programas_departamentos.json') as f:
   dados = json.load(f)

#resultado = requests.get(url = 'https://dados.fflch.usp.br/api/programas') 
#dados = resultado.json()
print(dados)
```

```python
from pandas.core.resample import f
lstDeps = dados [1] # secao de departamentos

import json
import sys


linhas_pesquisa = {}
for lst in lstDeps:
  departamento = lst['nome'] #aqui definimos algumas coisas, como pegar o departamento dentros dos dados, no caso 'nome'
  idLattes_docentes = lst['codpes_docentes']
  linhas_pesquisa[departamento] = {}
#idLattes_docentes = [3752790499157008, 4614382386818949, 9033064139380328,2569898005322201,2112694590578407] #os ids dos docentes estao como numero usp, porem procurar quando resolvido pelo id do lattes

  #pegando as informacoes de cada docente
  for idLattes in idLattes_docentes:
    url_request = f'https://dados.fflch.usp.br/api/programas/docente/{idLattes}'
    lattes = requests.get(url = url_request)
    print(url_request)
    
    try:
      lattes = lattes.json()
    except:
      print(lattes)
      sys.exit() 

    nome_docente = lattes['nome']
    lnPsq = lattes['linhas_pesquisa']
    linhas_pesquisa[departamento][nome_docente] = lnPsq

import json
print(json.dumps(linhas_pesquisa, sort_keys=True, indent=4))
print(linhas_pesquisa)
```
